#### 1 - Escreva uma função que gere uma senha aleatória para o usuário. O programa deve perguntar para o usuário quantas letras e quantos números desejará nessa senha, e a senha retornada deve ter aleatoriamente letras maiúsculas e minúsculas. A ordem entre letras e números não precisa ser aleatória.

```python
import random

def gera_senha(qtde_letras, qtde_numeros):
  senha_lista = []
  contador = 0
  while contador < qtde_numeros:
    senha_lista.append(str(random.randint(0, 9)))
    contador += 1

  letras = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'

  contador = 0
  while contador < qtde_letras:
    senha_lista.append(letras[random.randint(0, len(letras)-1)])
    contador += 1

  senha = ''
  while senha_lista:
    senha += senha_lista.pop(0)

  return senha
  
qtde_letras = int(input("Digite a quantidade de letras: "))
qtde_numeros = int(input("Digite a quantidade de numeros: "))
senha = gera_senha(qtde_letras, qtde_numeros)
print(senha)
```

#### 2 - Faça uma função que funcione como o jogo "pedra, papel, tesoura", onde um usuário joga contra o computador. A escolha do usuário é fornecida através do comando input(), enquanto que a do computador é escolhida aleatoriamente. O programa deve perguntar se o usuário quer jogar novamente.

#### Regras do jogo:

* papel vence pedra

* pedra vence tesoura

* tesoura vence papel

```python
import random

def input_validado(opcoes):
  opcao_usuario = input(f"O que deseja jogar? ({'/'.join(opcoes)}) ")
  while opcao_usuario not in opcoes:
    opcao_usuario = input(f"Opção inválida!\nO que deseja jogar? ({'/'.join(opcoes)}) ")
  else:
    return opcao_usuario


def confere_resultado(opcao_usuario, opcao_computador):
  if opcao_usuario == opcao_computador:
    return "Deu empate!"
  elif opcao_usuario == 'pedra':
    if opcao_computador == 'papel':
      return "Você perdeu!"
    else:
      return "Você ganhou!"
  elif opcao_usuario == 'papel':
    if opcao_computador == 'tesoura':
      return "Você perdeu!"
    else:
      return "Você ganhou!"
  elif opcao_usuario == 'tesoura':
    if opcao_computador == 'pedra':
      return "Você perdeu!"
    else:
      return "Você ganhou!"


def jokenpo():
  quer_jogar = 's'
  while quer_jogar == 's':
    opcoes = ['pedra', 'papel', 'tesoura']
    opcao_usuario = input_validado(opcoes)

    opcao_computador = opcoes[random.randint(0, len(opcoes)-1)]
    print(f"O computador jogou: {opcao_computador}")

    print(confere_resultado(opcao_usuario, opcao_computador))

    quer_jogar = input("Deseja jogar novamente? (s/n) ")
    while quer_jogar != 's' and quer_jogar != 'n':
      quer_jogar = input("Opção inválida!\nDeseja jogar novamente? (s/n) ")
  else:
    print("Obrigado por jogar!")


jokenpo()
```

#### 3 - Faça uma função que recebe uma lista com inteiros e devolva a lista sem os números divisíveis por 2 e/ou por 5.

```python
def limpa_divisiveis_2_5(lista):
  index = 0
  while index < len(lista):
    if lista[index] % 5 == 0 or lista[index] % 2 == 0:
      lista.pop(index)
      continue
    index += 1
  return lista

limpa_divisiveis_2_5([1, 2, 3, 5, 6, 7, 8])
```

#### 4 - Faça uma função que some todos os inteiros de uma lista de N dimensões. 

```python
def soma_aninhado(lista):
  soma = 0

  contador =  0
  while contador < len(lista):
    if type(lista[contador])  == int:
      soma += lista[contador]
    else:
      soma_parcial = soma_aninhado(lista[contador])
      print(soma_parcial)
      soma += soma_parcial

    contador += 1

  return soma

lista = [1, 4, 7, 10, 5, [5, 6, [4], [1, 10]], [6, 9, [4]]]
soma_aninhado(lista)
```

#### 5 - Faça uma função que receba uma lista de N dimensões e um valor e remova este valor de toda a estrutura, incluindo as sublistas.

```python
def elimina_elemento(lista, elemento):
  index = 0
  while index < len(lista):
    if type(lista[index]) == list:
      elimina_elemento(lista[index], elemento)
    elif lista[index] == elemento:
      lista.pop(index)
      continue
    index += 1

lista = [1, 5, 'a', True, [4, 5, 'b', ['d', True, 5], False]]
elimina_elemento(lista, 5)

print(lista)
```
